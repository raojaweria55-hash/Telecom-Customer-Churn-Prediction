```python
# ============================================
# TASK 02: Telecom Customer Churn
# ============================================

# --------- 1. Import Libraries ----------
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, confusion_matrix

from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier

sns.set_theme(style="ticks")

print("✅ Libraries Imported")

# --------- 2. Upload Dataset ----------
from google.colab import files
files.upload()

# --------- 3. Load Dataset ----------
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn (1).csv")

print("Dataset Shape:", df.shape)
df.head()

# --------- 4. Column Check ----------
print("Columns:\n", df.columns)

print("\nTarget Variable Distribution:")
print(df['Churn'].value_counts(normalize=True))


# --------- 5. Data Cleaning ----------

df.drop('customerID', axis=1, inplace=True)

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)

print("Missing values handled ✅")

 # --------- 6. Encode Target ----------
df['Churn'] = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)

print(df['Churn'].value_counts())
plt.figure(figsize=(6,4))
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title("Churn by Internet Service")
plt.show()
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='tenure', hue='Churn', bins=30)
plt.title("Tenure Distribution by Churn")
plt.show()

# --------- 7. Encoding ----------
df_encoded = pd.get_dummies(df, drop_first=True)

print("Encoded Shape:", df_encoded.shape)


# --------- 8. Split ----------
X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=1
)

print("Split Done ✅")


# --------- 9. Scaling ----------
scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print("Scaling Done ✅")

#---------10.Model Training---------

#1️⃣ Naive Bayes
nb = GaussianNB()
nb.fit(X_train, y_train)

#2️⃣ Support Vector Machine (SVM)
svm = SVC(kernel='rbf', probability=True)
svm.fit(X_train, y_train)

#3️⃣ Gradient Boosting (Advanced Tree Model)
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)

print("All models trained ✅")

  # --------- 11. Evaluation ----------
def evaluate(model, name):
    y_pred = model.predict(X_test)
    print(f"\n📌 {name}")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

evaluate(nb, "Naive Bayes")
evaluate(svm, "Support Vector Machine")
evaluate(gb, "Gradient Boosting")


# --------- 12. Feature Importance ----------
importance = gb.feature_importances_
features = X.columns

imp_df = pd.DataFrame({
    'Feature': features,
    'Importance': importance
}).sort_values(by='Importance', ascending=False)

imp_df.head(10)
plt.figure(figsize=(7,5))
sns.barplot(x='Importance', y='Feature', data=imp_df.head(8))
plt.title("Top Churn Drivers (Gradient Boosting)")
plt.show()


# --------- 13. Insights ----------
print("""
Key Observations:
• Customers with low tenure show higher churn
• Contract type strongly influences churn
• Service-related features impact retention
""")


# --------- 14. Save ----------
df_encoded.to_csv("telco_churn_version2_clean.csv", index=False)

print("✅ Version 2 dataset saved")


print("🎉 Telecom Churn Prediction – Alternate Version Completed!")
```
